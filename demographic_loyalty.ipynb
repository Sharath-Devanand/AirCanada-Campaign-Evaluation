```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_flight = pd.read_csv('data\Customer Flight Activity.csv')
df_customer = pd.read_csv('data\Customer Loyalty History.csv')
```

```python
df_flight = df_flight[['Loyalty Number','Total Flights','Points Accumulated','Month','Year']]
df_flight = df_flight.groupby(['Loyalty Number','Year','Month']).sum().reset_index()

df_customer = df_customer[['Loyalty Number','Salary','Marital Status','Education','Gender','Province']]


df_features = df_flight.merge(df_customer,on='Loyalty Number',how='left')
```

```python
df_features.head()
```

```python
### Total flights over 2017 based on Marital Status, Education, Gender and Province
```

```python
### Marital Status

df_marital_flights = df_features[['Year','Month','Marital Status','Total Flights']]
df_marital_flights = df_marital_flights.groupby(['Year','Month','Marital Status']).mean().reset_index()

plt.figure(figsize=(10,5))
sns.set_theme()
sns.barplot(data = df_marital_flights,x='Month',y='Total Flights',hue='Marital Status')
plt.title('Average number of flights taken in 2017 based on Marital Status')
plt.ylabel('Average Flights')
```

```python
### Loyalty Tier wise number of flights booked over time (solo and companion)

factor = 'Gender'

df_marital_flights = df_features[['Year','Month',factor,'Total Flights']]
df_marital_flights = df_marital_flights.groupby(['Year','Month',factor]).mean().reset_index()

plt.figure(figsize=(10,5))
sns.set_theme()
sns.barplot(data = df_marital_flights,x='Month',y='Total Flights',hue=factor)
plt.title('Average number of flights taken in 2017 based on Marital Status')
plt.ylabel('Average Flights')
```

```python
### Loyalty Tier wise number of flights booked over time (solo and companion)

factor = 'Province'

df_marital_flights = df_features[['Year','Month',factor,'Total Flights']]
df_marital_flights = df_marital_flights.groupby(['Year','Month',factor]).mean().reset_index()

plt.figure(figsize=(10,5))
sns.set_theme()
sns.lineplot(data = df_marital_flights,x='Month',y='Total Flights',hue=factor)
plt.title('Average number of flights taken in 2017 based on Marital Status')
plt.ylabel('Average Flights')
```

```python
### Loyalty Tier wise number of flights booked over time (solo and companion)

factor = 'Education'

df_marital_flights = df_features[['Year','Month',factor,'Total Flights']]
df_marital_flights = df_marital_flights.groupby(['Year','Month',factor]).mean().reset_index()

plt.figure(figsize=(10,5))
sns.set_theme()
sns.barplot(data = df_marital_flights,x='Month',y='Total Flights',hue=factor)
plt.title('Average number of flights taken in 2017 based on Marital Status')
plt.ylabel('Average Flights')
```
